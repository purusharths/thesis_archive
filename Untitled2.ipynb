```python
import parafields
import matplotlib.pyplot as plt
plt.rc('text', usetex=True)
plt.rc('font', family='serif', size=14)  # Adjust the size parameter as needed
```

```python
field = parafields.generate_field(
    cells=(256, 256), extensions=(1.0, 1.0), covariance="exponential", variance=2.0
)


a = field.evaluate()

plt.imshow(a, cmap="turbo")
```

```python
plt.imshow(a, cmap="turbo")
# plt.colorbar()
plt.tight_layout()
plt.xticks([])
plt.yticks([])
plt.savefig("delayed_response1.png", bbox_inches="tight")
```

```python
filedg = parafields.generate_field(
    cells=(256, 256),
    covariance="gaussian",
    variance=0.1,
    corrLength=0.1,
    periodic=True,
)
filedg.add_mean_trend_component(mean=1, variance=0.01)
```

```python
plt.imshow(filedg.evaluate(), cmap="turbo")
plt.xticks([])
plt.yticks([])
# plt.colorbar()
plt.tight_layout()
plt.savefig("gaussian__prob.pdf", bbox_inches="tight")
```

```python
filedg = parafields.generate_field(
    cells=(256, 256),
    covariance="gaussian",
    variance=0.2,
    corrLength=0.4,
    #periodic=True,
    autotune_embedding_factor=True
)


filedg.add_mean_trend_component(mean=1, variance=0.01)
plt.imshow(filedg.evaluate(), cmap="turbo")
```

```python
plt.imshow(filedg.evaluate(), cmap="turbo")
# plt.colorbar()
plt.xticks([])
plt.yticks([])
plt.tight_layout()
plt.savefig("gaussian_smooth___2.png", bbox_inches="tight")
```

```python
fld = parafields.generate_field(
    cells=(256, 256),
    covariance="exponential",
    variance=1.0,
    anisotropy="axiparallel",
    corrLength=[0.05, 0.2],
)
```

```python
plt.imshow(fld.evaluate(), cmap="turbo")
plt.colorbar()
plt.tight_layout()
plt.savefig("gaussian_isotropic1.pdf", bbox_inches="tight")
```
